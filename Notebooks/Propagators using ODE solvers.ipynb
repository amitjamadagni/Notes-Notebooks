```python
using QuBase
using QuDynamics
using PyPlot
```

#### Using numerical methods ode45, ode78, ode23s from ODE.jl to solve the below example.

```python
Hamiltonian = sigmax
# Initial State 
qv = normalize!(QuArray([0.5+0.1im, 0.2+0.2im]))
# Time steps 
tlist = 0.:0.1:2*pi
```

```python
# Creating an instance of QuPropagator, problem to be solved by ode45 from ODE.jl
qode45 = QuPropagator(Hamiltonian, qv, tlist, QuODE45())
# Putting the iterator methods into action, we plot the state
for (t, psi) in qode45
    # time and the state
    println("$t $(psi[1]) $(psi[2])")
    plot(t, Float64[real(psi[1])], "ro")
    plot(t, Float64[real(psi[2])], "ro")
end
```

```python
# Creating an instance of QuPropagator, problem to be solved by ode78 from ODE.jl
qode78 = QuPropagator(Hamiltonian, qv, tlist, QuODE78())
# Putting the iterator methods into action, we plot the state
for (t, psi) in qode78
    # time and the state
    println("$t $(psi[1]) $(psi[2])")
    plot(t, Float64[real(psi[1])], "ro")
    plot(t, Float64[real(psi[2])], "ro")
end
```

```python
# Creating an instance of QuPropagator, problem to be solved by ode45 from ODE.jl
qode23s = QuPropagator(Hamiltonian, qv, tlist, QuODE23s())
# Putting the iterator methods into action, we plot the state
for (t, psi) in qode23s
    # time and the state
    println("$t $(psi[1]) $(psi[2])")
    plot(t, Float64[real(psi[1])], "ro")
    plot(t, Float64[real(psi[2])], "ro")
end
```

```python
fs_a = Array(typeof(qv),0)
push!(fs_a, qv)
for i=2:length(tlist)
    push!(fs_a, expm(-im*Hamiltonian*(tlist[i]-tlist[1]))*fs_a[1])
end
for xe in fs_a
    println(coeffs(xe)[1]," ", coeffs(xe)[2])
end
plot(tlist, Float64[real(xe[1]) for xe in fs_a],"ro")
plot(tlist, Float64[real(xe[2]) for xe in fs_a],"ro")
```

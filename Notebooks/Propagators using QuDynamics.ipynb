```python
using QuBase
```

```python
using QuDynamics
```

```python
using PyPlot
```

```python
# Initial Conditions
Hamiltonian = sigmax
# Initial State 
qv = normalize!(QuArray([0.5+0.1im, 0.2+0.2im]))
# Time steps 
tlist = 0.:0.1:2*pi
```

```python
# Creating an instance of QuPropagator, problem to be solved by Euler Method
qe = QuPropagator(Hamiltonian, qv, tlist, QuEuler())
```

```python
# Evaluating the next state after the start state by using `Base.next` iterator method
start_state_qe = start(qe)
next_state_qe = next(qe, start_state_qe)
```

```python
# Putting the iterator methods into action, we plot the state
for (t, psi) in qe
    # time and the state
    println("$t $(psi[1]) $(psi[2])")
    plot(t, Float64[real(psi[1])], "ro")
    plot(t, Float64[real(psi[2])], "ro")
end
```

### Similarly using other numerical methods to solve the above example

```python
# Creating an instance of QuPropagator, problem to be solved by Crank Nicolson Method
qcn = QuPropagator(Hamiltonian, qv, tlist, QuCrankNicolson())
# Putting the iterator methods into action, we plot the state
for (t, psi) in qcn
    # time and the state
    println("$t $(psi[1]) $(psi[2])")
    plot(t, Float64[real(psi[1])], "ro")
    plot(t, Float64[real(psi[2])], "ro")
end
```

```python
# Creating an instance of QuPropagator, problem to be solved by Krylov subspace Method
qkr = QuPropagator(Hamiltonian, qv, tlist, QuKrylov())
# Putting the iterator methods into action, we plot the state
for (t, psi) in qkr
    # time and the state
    println("$t $(psi[1]) $(psi[2])")
    plot(t, Float64[real(psi[1])], "ro")
    plot(t, Float64[real(psi[2])], "ro")
end
```

```python
fs_a = Array(typeof(qv),0)
push!(fs_a, qv)
for i=2:length(tlist)
    push!(fs_a, expm(-im*Hamiltonian*(tlist[i]-tlist[1]))*fs_a[1])
end
for xe in fs_a
    println(coeffs(xe)[1], coeffs(xe)[2])
end
plot(tlist, Float64[real(xe[1]) for xe in fs_a],"ro")
plot(tlist, Float64[real(xe[2]) for xe in fs_a],"ro")
```
